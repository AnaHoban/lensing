```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
import csv
import seaborn as sns
```

```python
scratch = os.path.expandvars('$SCRATCH') + '/'
```

```python
master_cat = pd.read_csv(scratch+'master_catalogue.csv')
```

```python
master_cat = master_cat.dropna()
```

```python
#change those cols to strings
master_cat['TILE'].apply(str)
master_cat['BAND'].apply(str)

master_cat['TILE'] = [str(x)[:8] for x in master_cat['TILE']]
```

```python
master_cat
```

```python
print(master_cat['BAND'].value_counts())
#print(master_cat['TILE'].value_counts())
```

```python
from matplotlib.ticker import FormatStrFormatter
g = sns.countplot(x='TILE', data =  master_cat)
g.set_xticklabels(g.get_xticklabels(), rotation=60)
plt.show()
sns.countplot(x='BAND', data = master_cat)
```

```python
g = sns.countplot(x='TILE', data = master_cat[master_cat['BAND'] == 'u'])
plt.title('band = u')
g.set_xticklabels(g.get_xticklabels(), rotation=60)
plt.show()

g = sns.countplot(x='TILE', data = master_cat[master_cat['BAND'] == 'r'])
plt.title('band = r')
g.set_xticklabels(g.get_xticklabels(), rotation=60)
plt.show()
```

```python
sns.set_theme(style = 'darkgrid')
plot = sns.displot(master_cat, x='BAND', col = 'TILE')
#            binwidth=1,  facet_kws=dict(margin_titles=True))

plot.fig.set_figwidth(18)
plot.fig.set_figheight(6)
```

```python
#sns.set(rc={'figure.figsize':(18,6)})
sns.set_theme(style = 'darkgrid')
plot = sns.displot(master_cat, x='MAG_AUTO', col = 'TILE', row= 'BAND',
#           binwidth=1, facet_kws=dict(margin_titles=True))

plot.fig.set_figwidth(18)
plot.fig.set_figheight(12)
```

```python
sns.set_theme(style = 'darkgrid')
plot = sns.displot(master_cat, x= 'TILE', col = 'FLAGS', row= 'BAND',
           binwidth=1, facet_kws=dict(margin_titles=True))
plot.fig.set_figwidth(20)
plot.fig.set_figheight(12)
```

```python
sns.set_theme(style = 'darkgrid')
plot = sns.displot(master_cat, x= 'MAGERR_AUTO', col = 'TILE', row= 'BAND',
                   facet_kws=dict(margin_titles=True))
plot.fig.set_figwidth(28)
plot.fig.set_figheight(12)
```
